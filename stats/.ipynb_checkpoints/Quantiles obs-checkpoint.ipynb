```python
## general
import numpy as np
import datetime
import joblib
import copy
import xarray as xr
import os
import sys
import datetime as dt
import matplotlib as mpl
import math
import joblib
import numpy.ma as ma

## statistics
from scipy import stats
from scipy.stats import linregress

## plotting
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy.ma as ma
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import mplotutils as mpu
```

```python
# load the land mask as frac_l
dir_in_geo_dist = '/home/tristan/mesmer/data/'
frac_l = xr.open_mfdataset(dir_in_geo_dist + 'interim_invariant_lsmask_regrid.nc', combine='by_coords',decode_times=False)

frac_l_raw = np.squeeze(copy.deepcopy(frac_l.lsm.values))  #land-sea mask of ERA-interim bilinearily interpolated 

frac_l = frac_l.where(frac_l.lat>-60,0)  # remove Antarctica from frac_l field (ie set frac l to 0)

idx_l=np.squeeze(frac_l.lsm.values)>0.0 # idx_l = index land -> idex land #-> everything >0 we consider as land

lon_pc, lat_pc = mpu.infer_interval_breaks(frac_l.lon, frac_l.lat)  ## is this needed??
```

```python
dir_in_data_mod = '/home/tristan/mesmer/data/'

y_all = joblib.load(dir_in_data_mod+'y_all_fix.pkl')
y_all_mon = joblib.load(dir_in_data_mod+'y_all_mon_fix.pkl')
```

## Calculate percentiles for annual BEST data

```python
perc_obs_best = {}

perc_obs_best[2.5], perc_obs_best[5.0], perc_obs_best[50], perc_obs_best[95], perc_obs_best[97.5] = np.percentile(y_all,[2.5,5.0, 50,95,97.5],axis=0)
```

```python
print('5th pecentile min: ', perc_obs_best[5.0].min(),' & max: ',perc_obs_best[5.0].max())
print('50th pecentile min: ', perc_obs_best[50].min(),' & max: ',perc_obs_best[50].max())
print('95th pecentile min: ', perc_obs_best[95].min(),' & max: ',perc_obs_best[95].max())
```

```python
## mask the data 5th ###################################################
y_ma_5th = np.zeros(idx_l.shape)
y_ma_5th = ma.masked_array(y_ma_5th, mask=idx_l==False)
y_ma_5th[idx_l]=perc_obs_best[5.0]

print('5th pecentile min: ', y_ma_5th.min(),' & max: ',y_ma_5th.max())

######### 50th  ######################################################
y_ma_50th = np.zeros(idx_l.shape)
y_ma_50th = ma.masked_array(y_ma_50th, mask=idx_l==False)
y_ma_50th[idx_l]=perc_obs_best[50]

print('50th pecentile min: ', y_ma_50th.min(),' & max: ',y_ma_50th.max())

################### 95th #######################################
y_ma_95th = np.zeros(idx_l.shape)
y_ma_95th = ma.masked_array(y_ma_95th, mask=idx_l==False)
y_ma_95th[idx_l]=perc_obs_best[95]

print('95th pecentile min: ', y_ma_95th.min(),' & max: ',y_ma_95th.max())
```

## Calculate percentiles for monthly BEST data

```python
####### first for January #####################################
perc_obs_best_jan = {}

perc_obs_best_jan[2.5], perc_obs_best_jan[5.0], perc_obs_best_jan[50], perc_obs_best_jan[95], perc_obs_best_jan[97.5] = np.percentile(y_all_mon[0::12],[2.5,5.0, 50,95,97.5],axis=0)

##### next for July ###########################################
perc_obs_best_jul = {}

perc_obs_best_jul[2.5], perc_obs_best_jul[5.0], perc_obs_best_jul[50], perc_obs_best_jul[95], perc_obs_best_jul[97.5] = np.percentile(y_all_mon[6::12],[2.5,5.0, 50,95,97.5],axis=0)
```

```python
################### MASKING FOR JANUARY ######################################
## mask the data 5th ###################################################
y_ma_5th_jan = np.zeros(idx_l.shape)
y_ma_5th_jan = ma.masked_array(y_ma_5th_jan, mask=idx_l==False)
y_ma_5th_jan[idx_l]=perc_obs_best_jan[5.0]

print('5th pecentile min: ', y_ma_5th_jan.min(),' & max: ',y_ma_5th_jan.max())

######### 50th  ######################################################
y_ma_50th_jan = np.zeros(idx_l.shape)
y_ma_50th_jan = ma.masked_array(y_ma_50th_jan, mask=idx_l==False)
y_ma_50th_jan[idx_l]=perc_obs_best_jan[50]

print('50th pecentile min: ', y_ma_50th_jan.min(),' & max: ',y_ma_50th_jan.max())

################### 95th #######################################
y_ma_95th_jan = np.zeros(idx_l.shape)
y_ma_95th_jan = ma.masked_array(y_ma_95th_jan, mask=idx_l==False)
y_ma_95th_jan[idx_l]=perc_obs_best_jan[95]

print('95th pecentile min: ', y_ma_95th_jan.min(),' & max: ',y_ma_95th_jan.max())
```

```python
################### MASKING FOR JULY ######################################
## mask the data 5th ###################################################
y_ma_5th_jul = np.zeros(idx_l.shape)
y_ma_5th_jul = ma.masked_array(y_ma_5th_jul, mask=idx_l==False)
y_ma_5th_jul[idx_l]=perc_obs_best_jul[5.0]

print('5th pecentile min: ', y_ma_5th_jul.min(),' & max: ',y_ma_5th_jul.max())

######### 50th  ######################################################
y_ma_50th_jul = np.zeros(idx_l.shape)
y_ma_50th_jul = ma.masked_array(y_ma_50th_jul, mask=idx_l==False)
y_ma_50th_jul[idx_l]=perc_obs_best_jul[50]

print('50th pecentile min: ', y_ma_50th_jul.min(),' & max: ',y_ma_50th_jul.max())

################### 95th #######################################
y_ma_95th_jul = np.zeros(idx_l.shape)
y_ma_95th_jul = ma.masked_array(y_ma_95th_jul, mask=idx_l==False)
y_ma_95th_jul[idx_l]=perc_obs_best_jul[95]

print('95th pecentile min: ', y_ma_95th_jul.min(),' & max: ',y_ma_95th_jul.max())
```

## Make a plot

```python
### set plot params

params = {'legend.fontsize': 'xx-large',
         'axes.labelsize': 'x-large',
         'axes.titlesize':'xx-large',
         'xtick.labelsize':'x-large',
          'figure.figsize': (9,5),
          "font.family": "serif",
         'ytick.labelsize':'x-large',
         "font.weight":"normal",
         "axes.labelweight":"normal"}
plt.rcParams.update(params)
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
nrows = 3
ncols = 3

# Define the figure and each axis for the 3 rows and 3 columns
fig, ax = plt.subplots(nrows=nrows, ncols=ncols, 
                        subplot_kw={'projection': ccrs.Robinson(central_longitude=0)},
                        figsize=(20,15))

# axs is a 2 dimensional array of `GeoAxes`.  We will flatten it into a 1-D array
ax=ax.flatten()

## set bounds and norms
bounds = np.arange(-2.5,4.5,0.5)
norm = colors.BoundaryNorm(boundaries=bounds, ncolors=256)
cmap = plt.cm.get_cmap('RdYlBu_r')

bounds_jan = np.arange(-35,16,5)
norm_jan = colors.BoundaryNorm(boundaries=bounds_jan, ncolors=256)

bounds_jul = np.arange(-15,35,5)
norm_jul = colors.BoundaryNorm(boundaries=bounds_jul, ncolors=256)

################################### PLOT FOR ANNUAL TEMPS FIRST ######################################################
### 5th percentile
mesh_1=ax[0].pcolormesh(lon_pc, lat_pc, y_ma_5th, cmap=cmap, norm=norm, transform=ccrs.PlateCarree(),rasterized=True)
ax[0].coastlines()
#ax[0].add_feature(cfeature.OCEAN)
ax[0].set_title('5 %')

#### 50th percentile
mesh_1=ax[1].pcolormesh(lon_pc, lat_pc, y_ma_50th, cmap=cmap, norm=norm, transform=ccrs.PlateCarree(),rasterized=True)
ax[1].coastlines()
#ax[1].add_feature(cfeature.OCEAN)
ax[1].set_title('50 %')

#### 95th percentile
mesh_1=ax[2].pcolormesh(lon_pc, lat_pc, y_ma_95th, cmap=cmap, norm=norm, transform=ccrs.PlateCarree(),rasterized=True)
ax[2].coastlines()
#ax[2].add_feature(cfeature.OCEAN)
ax[2].set_title('95 %')

plt.text(-0.08,0.35,s='Annual',transform=ax[0].transAxes, ha='center',rotation=90, fontsize=16)

cbar_1=fig.colorbar(mesh_1,ax=ax[0:3],orientation='horizontal',ticks=([-2.0, -1.0, 0.0, 1.0, 2.0, 3.0, 4.0]), location='bottom', pad=0.05, shrink=0.35, aspect=35)
#pad=30,aspect=50
cbar_1.set_label('Temperature [°C]') 

################################### PLOT FOR JAN TEMPS ######################################################
### 5th percentile
mesh_2=ax[3].pcolormesh(lon_pc, lat_pc, y_ma_5th_jan, cmap=cmap, norm=norm_jan, transform=ccrs.PlateCarree(),rasterized=True)
ax[3].coastlines()
#ax[0].add_feature(cfeature.OCEAN)
ax[3].set_title('5 %')

#### 50th percentile
mesh_2=ax[4].pcolormesh(lon_pc, lat_pc, y_ma_50th_jan, cmap=cmap, norm=norm_jan, transform=ccrs.PlateCarree(),rasterized=True)
ax[4].coastlines()
#ax[1].add_feature(cfeature.OCEAN)
ax[4].set_title('50 %')

#### 95th percentile
mesh_2=ax[5].pcolormesh(lon_pc, lat_pc, y_ma_95th_jan, cmap=cmap, norm=norm_jan, transform=ccrs.PlateCarree(),rasterized=True)
ax[5].coastlines()
#ax[2].add_feature(cfeature.OCEAN)
ax[5].set_title('95 %')

plt.text(-0.08,0.35,s='January',transform=ax[3].transAxes, ha='center',rotation=90, fontsize=16)

cbar_2=fig.colorbar(mesh_2,ax=ax[3:6],orientation='horizontal',ticks=([-35, -25, -15, -5, 5, 15]), location='bottom', shrink=0.35,pad=0.05,aspect=35)
cbar_2.set_label('Temperature [°C]') 


################################### PLOT FOR JULY TEMPS ######################################################
### 5th percentile
mesh_3=ax[6].pcolormesh(lon_pc, lat_pc, y_ma_5th_jul, cmap=cmap, norm=norm_jul, transform=ccrs.PlateCarree(),rasterized=True)
ax[6].coastlines()
#ax[0].add_feature(cfeature.OCEAN)
ax[6].set_title('5 %')

#### 50th percentile
mesh_3=ax[7].pcolormesh(lon_pc, lat_pc, y_ma_50th_jul, cmap=cmap, norm=norm_jul, transform=ccrs.PlateCarree(),rasterized=True)
ax[7].coastlines()
#ax[1].add_feature(cfeature.OCEAN)
ax[7].set_title('50 %')

#### 95th percentile
mesh_3=ax[8].pcolormesh(lon_pc, lat_pc, y_ma_95th_jul, cmap=cmap, norm=norm_jul, transform=ccrs.PlateCarree(),rasterized=True)
ax[8].coastlines()
#ax[2].add_feature(cfeature.OCEAN)
ax[8].set_title('95 %')

plt.text(-0.08,0.35,s='July',transform=ax[6].transAxes, ha='center',rotation=90, fontsize=16)

cbar_3=fig.colorbar(mesh_3,ax=ax[6:9],orientation='horizontal', ticks=([-15, -5, 5, 15, 25]),shrink=0.35, location='bottom', pad=0.05, aspect=35)
cbar_3.set_label('Temperature [°C]') 

#plt.subplots_adjust(wspace=0.1, hspace=-0.5)

fig.tight_layout()
```
